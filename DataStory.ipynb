```python
import pandas as pd

from matplotlib import pyplot as plt

import seaborn as sns

import numpy as np
```

```python
df = pd.read_parquet('Focus_playlist.parquet')
```

```python
hist = sns.histplot(stat="count", multiple="stack",
             x=(df.Duration_total_seconds/60), kde=False, hue=df.super_genre,
             element="bars", legend=False, bins=np.arange(0,12), hue_order=['Other', 'Jazz/Funk', 'Soundtrack', 'Ambiance/Chill', 'EDM/Drum-n-Bass', 'Classical'])
plt.xlabel('Song length (minutes)')
plt.title('Genres by song length in my Spotify "Focus" playlist')
hist.legend(title='Genre', labels=['Classical', 'EDM/Drum-n-Bass', 'Ambiance/Chill', 'Soundtrack', 'Jazz/Funk', 'Other'])
plt.savefig('DataStorySongLengthGenre.png')
plt.show()
```
